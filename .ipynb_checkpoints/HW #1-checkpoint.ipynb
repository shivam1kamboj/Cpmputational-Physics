# Problem 1: Newman 2.6: Planetary orbits

(a) According to the given problem $l_1$ and $v_1$  are distance and velocity at perihelion, and $l_2$ and $v_2$ are distance and velocity at aphelion. 
Show that $v_2$ is smaller root of the equation
$\begin{equation}v_2^2-\frac{2GM}{v_1l_1}v_2-\left[v_1^2-\frac{2GM}{l_1}\right]=0\end{equation}$\
$\textbf{Solution}$: Rearranging the equation: $$v_2^2-\frac{2GM}{v_1l_1}v_2=v_1^2-\frac{2GM}{l_1}$$
Substituting: $$v_1=\frac{l_2}{l_1}v_2$$ 
$$v_2^2-\frac{2GM}{l_2}=\left(\frac{l_2}{l_1}v_2\right)^2-\frac{2GM}{l_1}$$
Now rearranging again: $$\left[1-\left(\frac{l_2}{l_1}\right)^2\right]v_2^2=2GM\left(\frac{1}{l_2}-\frac{1}{l_1}\right)$$
$$\left[\frac{l_1^2-l_2^2}{l_1^2}\right]v_2^2=2GM\left(\frac{l_1-l_2}{l_1l_2}\right)$$
$$\left[\frac{(l_1-l_2)(l_1+l_2)}{l_1^2}\right]v_2^2=2GM\left(\frac{l_1-l_2}{l_1l_2}\right)$$
$$v_2^2=2GM\left(\frac{l_1}{l_2(l_1+l_2)}\right)$$\
$$\boxed{v_2=\sqrt{\frac{2 GM l_1}{l_2(l_1+l_2)}}}$$
Using $v_1=\frac{l_2}{l_1}v_2$ in the above equation we get:
$$\frac{l_1}{l_2}v_1=\sqrt{\frac{2 G M l_1}{l_2(l_1+l_2)}}$$\
$$v_1=\sqrt{\frac{2G M l_1 l_2^2}{l_1^2l_2(l_1+l_2)}}$$\
$$\boxed{v_1=\sqrt{\frac{2G M l_2}{l_1(l_1+l_2)}}}$$
$\textbf{Comparing $v_2$ and $v_1$ since $\frac{l_1}{l_2} < \frac{l_2}{l_1}$ as $l_2$ is larger (distance at aphelion) than $l_1$, therefore, $v_2$ is always smaller than $v_1$}$

(b) Now using quadratic formula to calculate $v_2$ form the given equation $$\begin{equation}v_2^2-\frac{2GM}{v_1l_1}v_2-\left[v_1^2-\frac{2GM}{l_1}\right]=0\end{equation}$$ gives
$$v_2=\frac{\frac{2GM}{v_1l_1}\pm\sqrt{\left(\frac{2GM}{v_1l_1}\right)^2-4\left[v_1^2-\frac{2GM}{l_1}\right]}}{2}$$

```python
M = 1.9891e30
def Orbital_parameters():
    v_1=float(input("Velocity at perihelion (m/s):"))
    l_1=float(input("distance at perihelion (m):"))
    p= -2*G*M/(v_1*l_1)
    c=-(v_1**2-(2*G*M)/l_1)
    v_2=(-p-np.sqrt(p**2-4*c))/2 #quadratic formula
    l_2=(l_1*v_1)/v_2
    a=(l_1+l_2)/2
    b=np.sqrt(l_1*l_2)
    T=(2*np.pi*a*b)/(l_1*v_1)
    conversion=365*24*3600
    t=T/conversion
    e=(l_2-l_1)/(l_2+l_1)
    print("velocity at aphelion=", v_2,'m/s'),print("distance from sun at aphelion=", l_2,'m'),print("Time period=", T,"s"),print("eccentricity=", e),print("time period in years=",t,"years")    
```

# Problem 3

Let's first create meshgrid using linear space function for both x and y by putting the limit -2 and 2

```python
n=1000   #number of iteration
x_grid = np.linspace(-2,2,n)
y_grid = np.linspace(-2,2,n)
x,y = np.meshgrid(x_grid,y_grid)
```

Now as given in the problem $c=x+iy$, let's create an empty array with same size and shape like c. Then simply iterate for the value n which number of elements in linear space

```python
c=(x+1j*y)
z=np.zeros_like(c)
for i in range(n):      
    z=z*z+c                         #here I never needed to use the constraint abs(z)>=2 as my linear space is 
                                    #already between -2 and 2. So all the values are within the limit automatically       
```

```python
print(np.shape(z))
print(np.size(z))
np.shape(np.abs(z))
```

Now plotting the absolute value of z using imshow

```python
fig1 = plt.figure(figsize=(10,10))
beetle = plt.imshow(np.abs(z))
cbar = fig1.colorbar(beetle)
cbar.set_label('density',fontsize=25)
```

```python
fig2 = plt.figure(figsize=(10,10))
beetle2 = plt.imshow(np.abs(z),cmap='magma')
cbar = fig2.colorbar(beetle2)
cbar.set_label('density',fontsize=25)
```
