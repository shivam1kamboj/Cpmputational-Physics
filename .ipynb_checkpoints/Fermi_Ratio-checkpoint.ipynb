### Demonstration of how to evaluate and plot Fermi-Dirac ratio from dielectric matrix

\begin{align}
\sum_{i \ne j} M_{ij} F \left( \epsilon_i, \epsilon_j \right) \\
F \left( \epsilon_i, \epsilon_j \right) = \frac{f \left( \epsilon_i \right) - f \left( \epsilon_j \right)}{\epsilon_i - \epsilon_j} \\
f \left( \epsilon \right) = \frac{1}{1 + e^{\left(\epsilon-\epsilon_{\rm F}\right)/k_B T}}
\end{align}

Ref: Jack Deslippe, Georgy Samsonidze, David A. Strubbe, Manish Jain, Marvin L. Cohen, and Steven G. Louie, "BerkeleyGW: A Massively Parallel Computer Package for the Calculation of the Quasiparticle and Optical Properties of Materials and Nanostructures," *Comput. Phys. Commun.* 183, 1269 (2012)

```python
from math import exp
#from scipy.constants import k
#https://docs.scipy.org/doc/scipy/reference/constants.html
k_B = 8.617333262e-5 
# eV K-1, https://physics.nist.gov/cgi-bin/cuu/Value?kev

fermi_energy = 0.0
temperature = 300
# unit: K
def fermi_dirac(energy):
    return 1/(1+exp((energy-fermi_energy)/(k_B*temperature)))
```

\begin{align}
\lim_{\Delta \rightarrow 0} \frac{f \left( \epsilon + \Delta \epsilon \right) - f \left( \epsilon \right)}{\Delta \epsilon} = f' \left( \epsilon \right) = -\frac{1}{k_B T} \frac{e^{\left(\epsilon-\epsilon_{\rm F}\right)/k_B T}}{\left( 1 + e^{\left(\epsilon-\epsilon_{\rm F}\right)/k_B T} \right)^2}
\end{align}

```python
def fermi_dirac_derivative(energy):
    return -exp((energy-fermi_energy)/(k_B*temperature))/((k_B*temperature)*(1+exp((energy-fermi_energy)/(k_B*temperature)))**2)
```

```python
from pylab import figure,legend
delta_list = [0.5, 0.1, 0.01, 1e-4, 1e-10, 1e-12, 1e-13, 1e-14, 1e-15, 1e-16, 1e-17, 1e-18, 1e-19]
figure(num=1, figsize=(12, 10))
# https://stackoverflow.com/questions/332289/how-do-you-change-the-size-of-figures-drawn-with-matplotlib
offset = 0

for delta in delta_list:
    xpoints = []
    ypoints = []
    # Note: indentation is crucial in python, part of syntax, not just "style" -- that determines which lines
    # are part of which nested loop here.
    for x in linspace(-0.2,0.2,100):
        xpoints.append(x)
        ypoints.append((fermi_dirac(x+delta)-fermi_dirac(x))/delta + offset)
    plot(xpoints,ypoints,label="delta = " + str(delta))
    # this is creating a string with text plus ("concatenated") with a number
    offset = offset - 1.5

xpoints = []
ypoints = []
for x in linspace(-0.2,0.2,100):
    xpoints.append(x)
    ypoints.append(fermi_dirac_derivative(x)+offset)
plot(xpoints,ypoints,label="derivative")
# https://stackoverflow.com/questions/19125722/adding-a-legend-to-pyplot-in-matplotlib-in-the-simplest-manner-possible
        
xlabel("E - E_F / eV")
ylabel("Ratio")
ylim(-40,2)
legend()
show()
```
