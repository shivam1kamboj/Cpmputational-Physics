```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
```

## Jacobi Method

```python
start=time()
# Constants
M = 100         # Grid squares on a side
V = 1.0         # Voltage at top wall
target = 1e-6   # Target accuracy

# Create arrays to hold potential values
phi1 = np.zeros([M+1,M+1],float)
phi1[0,:] = V
phiprime = np.empty([M+1,M+1],float)

# Main loop
delta = 1.0
iterations=0
while delta>target:

    # Calculate new values of the potential
    for i in range(M+1):
        for j in range(M+1):
            if i==0 or i==M or j==0 or j==M:
                phiprime[i,j] = phi1[i,j]
            else:
                phiprime[i,j] = (phi1[i+1,j] + phi1[i-1,j]+ phi1[i,j+1] + phi1[i,j-1])/4

    # Calculate maximum difference from old values
    delta = np.max(abs(phi1-phiprime))

    # Swap the two arrays around
    phi1,phiprime = phiprime,phi1
    iterations+=1
end=time()
print('run time in this cell is ',end-start,'seconds')
    # Make a plot
plt.imshow(phi1)
plt.gray()
plt.show()
print(delta)
print(iterations)
```
