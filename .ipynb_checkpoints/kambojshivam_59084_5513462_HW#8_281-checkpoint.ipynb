# Problem#1. Newman 8.3: The Lorenz equations

# Grade 10/10

One of the most celebrated sets of differential equations in physics is the Lorenz equations:

$\large\frac{dx}{dt}$ $=\sigma(y-x)$, $\hspace{10mm}\large\frac{dy}{dt}$ $=rx-y-xz$,  $\hspace{10mm}\large\frac{dz}{dt}$ $=xy-bz$

## Part a

$\textbf{Question}:$ Write a program using the Runge-Kutta fourth-order method to solve the Lorenz equations for the case $\sigma= 10$, $r = 28$, and $b = 83$ in the range from $t = 0$ to $t = 50$ with initial conditions $(x, y, z) = (0, 1, 0)$ , using $N = 10000$ points. 

$\textbf{Solution}:$

```python
### Solving Lorenz equations
name='RK4'
def f(r,t): ###Though f is not a function of t but as suggested in textbook keeping t won't change the answer
    sig = 10  ## value of sigma is given
    R = 28    ## value of r is given, here I named constant 'r' as 'R'
    B = 8/3   ## value of b is given, and I named constant 'b' as 'B'
    x,y,z = r[0], r[1], r[2] 
    fx = sig*(y - x)
    fy = R*x - y -x*z
    fz = x*y - B*z
    return np.array([fx,fy,fz],float)

a,b,N = 0.0,50,10000
h = (b-a)/N

t = np.arange(a,b,h)  ### independent variable t

r = np.zeros((len(t),3)) ### vector r = ([x,y,z]): started with all zeros 

r[0,0],r[0,1],r[0,2] = 0,1,0  ### initial values given in the problem (x,y,z)=(0,1,0)

### evaluating r using RK4 in t:[0,50]
for i in range(len(t)-1):
    k1 = h*f(r[i],t[i])
    k2 = h*f(r[i]+0.5*k1,t[i]+0.5*h)
    k3 = h*f(r[i]+0.5*k2,t[i]+0.5*h)
    k4 = h*f(r[i]+k3,t[i]+h)
    r[i + 1] = r[i] + (k1+2*k2+2*k3+k4)/6

x,y,z = r[:,0],r[:,1],r[:,2]

### plotting z(t)   
fig = plt.figure(figsize=(12,8))
plt.xlabel("t",fontsize=15)
plt.ylabel("$z(t)$",fontsize=15)               
plt.plot(t,z,label='z(t)')
plt.title('Solution of Lorenz equation using '+name+' with N= '+str(N),fontsize=20)
plt.legend(loc='best',fontsize=15)

### Now plotting the results x(t), y(t), and z(t)   
fig = plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.xlabel("t",fontsize=15)
plt.ylabel("$x(t)$",fontsize=15)                 
plt.plot(t,x,label='x(t)')       #### x(t)
plt.legend(loc='best',fontsize=15)

plt.subplot(1,2,2)
plt.xlabel("t",fontsize=15)
plt.ylabel("$y(t)$",fontsize=15)               
plt.plot(t,y,label='y(t)')       #### y(t)
plt.legend(loc='best',fontsize=15)
plt.show()
```

# Problem#2 Verlet method

# Grade 10/10

## Part a

$\textbf{Question:}$ Write a program to calculate the orbit of the Earth using the Verlet method, Eqs. (8.77) and (8.78), with a time-step of h = 1 hour. Make a plot of the orbit, showing several complete revolutions about the Sun. The orbit should be very slightly, but visibly, non-circular.

$\textbf{Solution:}$

```python
##value of x only at time t plus integer multiples of h and the value of v only at half-integer multiples. 
##We never need to calculate v and x at any of the integer points and at the half integer points respectively.
from scipy.constants import G

M=1.9891e30

name='Verlet method'
def f(r):   ## vector r
    x,y = r[0],r[1]
    r0 = np.sqrt(x**2+y**2)
    return np.array([-G*M*x/r0**3,-G*M*y/r0**3],float)

N = int(10*365*24) ### length of r_vector 10 years  
h = 3600             ###Step size 1-hour (converted in seconds)

v = np.zeros((2*N,2))    ### velocity vector v = v([vx,vy]) : started with all zeros 
r = np.zeros((N,2))      ### position vector r = ([x,y]): started with all zeros 

### initial values given in the problem 
v_initial = np.array([0,3.0287e4])       
r_initial = np.array([-1.4710e11,0])

v[0] = v_initial
r[0] = r_initial  

### evaluating r using Verlet method

v[1] = v[0]+0.5*h*f(r[0])  ##v(t+1/2h)=v(t)+1/2hf(r,t) first step for v using Euler's method to get started

## This is from eq 8.77 and 8.78 , in this v is computed for both half and integer multiple of h which require v
## twice the length of r as v is divided in h/2 stepsize for the same period
for i in range(N-1):
    r[i+1] = r[i]+h*v[2*i+1]
    k = h*f(r[i+1])
#     v[2*(i+1)] = v[2*i+1]+0.5*k   ## doesn't require this value for r
    v[2*(i+1)+1] = v[2*i+1]+k

xE,yE= r[:,0],r[:,1]

### Now plotting the results
fig = plt.figure(figsize=(10,8))
plt.plot(xE,yE, label = "Earth's trajectory")  ### Orbit of earth
plt.plot(0,0,'ro', markersize=10,label = "Sun's positon")                  
plt.plot(-1.4710e11,0,'bo',markersize=5, label = "Earth's initial positon")  # dark blue marker
plt.xlabel("$x$",fontsize=15)
plt.ylabel('$y$',fontsize=15)
plt.title("Earth's orbit using "+name ,fontsize=18)
plt.legend(loc='upper left',fontsize=10)
plt.show()
```

## Part b

$\textbf{Question:}$ Modify your program to calculate both potential $(−GMm/r)$, and kinetic energy $(\frac{1}{2}mv^2)$ of these quantities at each step, along with their sum (which is the total energy), and make a plot showing all three as a function of time on the same axes. You should find that the potential and kinetic energies vary visibly during the course of an orbit, but the total energy remains constant.

$\textbf{Solution:}$

```python
## vector r and initial values of position and velocity are taken from previous cell

m = 5.9722e24   ### mass of the planet Earth

def PE(r):       #Potential Energy
    x,y = r[0],r[1]
    r0 = np.sqrt(x**2+y**2)
    return -G*M*m/r0    ### Computed in S.I unit G in m^3/kgs^2 So PE is in joules

def KE(v):   ##Kinetic Energy
    vx,vy = v[0],v[1]
    v0 = np.sqrt(vx**2+vy**2)   ## speed v0
    return 0.5*m*v0**2

Potential = np.zeros((N,1))    
Kinetic = np.zeros((N,1))  
Total_energy = np.zeros((N,1)) 

Potential[0] = PE(r[0])
Kinetic[0] = KE(v[0])
Total_energy[0] = Potential[0] + Kinetic[0]

# ### evaluating r,v using Verlet method

v[1] = v[0]+0.5*h*f(r[0])  ##v(t+1/2h)=v(t)+1/2hf(r,t) first step using Euler's method to get started


for i in range(N-1):
    r[i+1] = r[i]+h*v[2*i+1]
    k = h*f(r[i+1])
    v[2*(i+1)] = v[2*i+1]+0.5*k   ## Need this value now for kinetic energy
    v[2*(i+1)+1] = v[2*i+1]+k
    Potential[i+1] = PE(r[i+1])
    Kinetic[i+1] = KE(v[2*(i+1)])
    Total_energy[i+1] = Potential[i+1] + Kinetic[i+1]


Time = np.arange(0,N,1)
### Now plotting the results
fig = plt.figure(figsize=(10,8))
plt.plot(Time, Potential, label = "Potential Energy")  
plt.plot(Time, Kinetic,label = "Kinetic Energy")                  
plt.plot(Time, Total_energy , label = "Total Energy")
plt.xlabel("Time, $(hours)$",fontsize=15)
plt.ylabel('$Energy$',fontsize=15)
plt.title("Earth's PE, KE, Total energy for 10 complete cycles " , fontsize =20,loc='right')
plt.legend(loc='best',fontsize=15)
plt.show()
```

## Part c
$\textbf{Question:}$ Now plot the total energy alone without the others and you should be able to see a slight variation over the course of an orbit. Because you’re using the Verlet method, however, which conserves energy in the long term, the energy should always return to its starting value at the end of each complete orbit.

$\textbf{Solution:}$

```python
fig = plt.figure(figsize=(10,8))
plt.plot(Time, Total_energy ,color='green')
plt.xlabel("Time, $(hours)$",fontsize=15)
plt.ylabel('Total Energy',fontsize=15)
plt.title("Earth's Total energy for 10 complete cycles " , fontsize =20,loc='right',color ='red')
plt.show()
```

## Part a

$\textbf{Question:}$ Consider the one-dimensional, time-independent Schrodinger equation in a harmonic (i.e., quadratic) potential $V(x) = V_0 \frac{x^2}{a^2}$, where $V_0$ and $a$ are constants. Write down the Schrodinger equation for this problem and convert it from a second-order equation to two first-order ones, as in Example 8.9. Using the shooting method as modified for eigenvalue problems, write a program or modify the one from Example 8.9, to find the energies of the ground state and the first two excited states for these equations when m is the electron mass, $V_0 = 50 eV$, and $a = 10^{−11}$ m.

$\textbf{Solution:}$ 

Schrodinger equation for this problem would be:

$$-\frac{\hbar^2}{2m}\frac{d^2\psi}{dx^2}+V_0\frac{x^2}{a^2}\psi(x)=E\psi(x)$$

Now converting it from a second-order equation to two first-order ones:

$$\frac{d\psi}{dx}=\phi\hspace{15mm}\frac{d\phi}{dx}=\frac{2m}{\hbar^2}\biggr(V_0\frac{x^2}{a^2}-E\biggr)\psi$$



```python
# Constants
from scipy.constants import hbar, m_e as m, e
a = 1e-11     # Bohr radius

## Writing a large but finite interval for wavefunction
x0 = -10*a   
xL = 10*a
N = 1000
h=(xL-x0)/1000

# Potential function
V_0=50*e ### eV to joule
def V(x):
    return V_0*x**2/a**2

def f(r,x,E):
    psi = r[0]
    phi = r[1]
    fpsi = phi
    fphi = (2*m/hbar**2)*(V(x)-E)*psi
    return np.array([fpsi,fphi],float)

# Calculate the wavefunction for a particular energy
def solve(E):
    psi = 0.0
    phi = 1.0
    r = np.array([psi,phi],dtype=object)

    for x in np.arange(x0,xL,h):
        k1 = h*f(r,x,E)
        k2 = h*f(r+0.5*k1,x+0.5*h,E)
        k3 = h*f(r+0.5*k2,x+0.5*h,E)
        k4 = h*f(r+k3,x+h,E)
        r += (k1+2*k2+2*k3+k4)/6
    return r[0]
```

```python
E=np.linspace(100*e,1000*e,100)
psi_L=[]
for EE in E:
    E_solve=solve(EE)
    psi_L.append(E_solve)
```

```python
### Now plotting the results
fig = plt.figure(figsize=(10,8))
plt.xlabel("$E$",fontsize=15)
plt.ylabel('$\psi(L)$',fontsize=15)
plt.title("$\psi(L)$ vs E" , fontsize =20)
plt.plot(E,psi_L)
plt.axhline(y= 0, color = 'r', linestyle = '-')
plt.xlim(100*e,1000*e)
plt.show()
```

```python
# Main program to find the ground state energy using the secant method
E1 = 115*e
E2 = e
psi2 = solve(E1)

target = e/1000
while abs(E1-E2)>target:
    psi1,psi2 = psi2,solve(E2)
    E1,E2 = E2,E2-psi2*(E2-E1)/(psi2-psi1)

E0=E2/e  ## ground state energy
print("E =",E0,"eV")
```

```python
# Main program to find the first excited state energy using the secant method
E1 = 400*e
E2 = e
psi2 = solve(E1)

target = e/1000
while abs(E1-E2)>target:
    psi1,psi2 = psi2,solve(E2)
    E1,E2 = E2,E2-psi2*(E2-E1)/(psi2-psi1)

E_1=E2/e  ## first excited state energy
print("E =",E_1,"eV")
```

```python
# Main program to find the second excited state energy using the secant method
E1 = 500*e
E2 = e
psi2 = solve(E1)

target = e/1000
while abs(E1-E2)>target:
    psi1,psi2 = psi2,solve(E2)
    E1,E2 = E2,E2-psi2*(E2-E1)/(psi2-psi1)

E_2=E2/e  ## second excited state energy
print("E =",E_2,"eV")
```

### Calculating first two excited states by using $E_n=\left(n+\frac{1}{2}\right)\hbar\omega$
As ground state energy is $$E_0=\frac{1}{2}\hbar\omega$$

Therefore, 
$$E_n=2\left(n+\frac{1}{2}\right)E_0$$

```python
E_first=2*(1+1/2)*E0
E_second=2*(2+1/2)*E0
print('Energy of the ground state is ',E0)
print('Energy of the first excited state is ',E_first)
print('Energy of the second excited state is ',E_second)
```
