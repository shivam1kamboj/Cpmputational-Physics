# Problem#2: Newman 9.4: Thermal diffusion in the Earths crust

# Grade 10/10

```python
# Constants
L = 20      # Depth in meters
D = 0.1      # Thermal diffusivity in m^2*day^-1
N = 100     # Number of divisions in grid
a = L/N       # Grid spacing
h = 0.01       # Time-step in days

AA=10
BB=12
def T0(t):
    return AA+BB*np.sin(2*np.pi*t/365)

Ninth_year=9*365  ## 9 years

c = h*D/(a**2)
```

```python
### checking stability condition i.e. a^2/h>2D
a*a/h,2*D,a*a/h>2*D,Ninth_year
```

```python
# Main Function to find the temperature values
T = np.zeros(N+1,float)
T[1:N]=10
T[N] = 11
def find_temp(T_array,t,Time):
    while t<Time:

        # Calculate the new values of T
        T[0] = T0(t)
        T[1:N] = T[1:N] + c*(T[0:N-1]+T[2:N+1]-2*T[1:N])
        t += h
    return T

T_ninth=find_temp(T_array=T,t=0,Time=3285)
plt.plot(np.linspace(0,L,N+1),T_ninth);
```

```python
fig = plt.figure(figsize=(10,8))
plt.title('Temperature change as a function of depth ',fontsize=15)
t_max=[3376,3467,3558, 3650]
for n in range(len(t_max)):
    t=3285
    T=find_temp(T_ninth,t,t_max[n])
    plt.plot(np.linspace(0,L,N+1),T,label='Nineth year + '+str(n+1)+' Quarter year')
    t=t_max[n]
plt.legend()
plt.xlabel("x,(m)")
plt.ylabel("T,(C)")
plt.show()
```

# Problem#3 Newman 9.8: The Schr odinger equation and the CrankNicolson method

# Grade 9/10
# Comment: x and y axes are wrong in c) -- this is a function of space not time. Physics interpretation is incomplete -- it is not just spreading, it is propagating to the right, and reflecting.

```python
from scipy.constants import hbar, m_e as M, e
from banded import banded
##choosing timestep h
h = 1e-18

L = 1e-8

N = 1000 # Grid slices
a = L/N

a1 = 1 + h*hbar*1j/(2*M*a**2)
a2 = -h*hbar*1j/(4*M*a**2)

b1 =  1 - h*hbar*1j/(2*M*a**2)
b2 =  h*hbar*1j/(4*M*a**2)

psi = np.zeros(N,complex)

x0 = L/2
sigma = 1e-10
k = 5e10
##defining psi0
def psi0(x):
    return np.exp(-(x-x0)**2/(2*sigma**2))*np.exp(1j*k*x)

x = np.linspace(0,L,N)

psi[:] = psi0(x)

## Defining B matrix because I will be using matrix multiplication to get v
def B_matrix(N):
    E=np.zeros((N,N))
    D0=np.tile(np.array([b1]), N)
    D1=np.tile(np.array([b2]), N-1)
    return E+np.diag(D0,k=0)+np.diag(D1,k=-1)+np.diag(D1,k=1)
```

```python
plt.imshow(abs(B_matrix(1000))[0:10,0:10]);   ##Pictorial representation of B_matrix just to verify
```

```python
B=B_matrix(1000)
#Bounday conditions 
psi[0]=psi[N-1]=0
A_band = np.zeros((3,N),complex)  ## Banded matrix

A_band[0,1:] = a2
A_band[1,:] = a1
A_band[2,:-1] = a2
```

```python
### first step of CrankNicolson method
v1=np.matmul(B,psi)    ###Calculated vector v
psi1=banded(A_band,v1,1,1)   ### calculated psi(t+h)

### Now calculating for first 500 wavefunction
psi_t=np.empty((500,N),complex) #Calculating psi values for successive timesteps and storing in a new array
psi_t[0]=psi

for i in range(1,len(psi_t)):
    v=np.matmul(B,psi_t[i-1])
    psi_t[i]=banded(A_band,v,1,1)
    psi_t[:,0]=psi_t[:,N-1]=0 

Psi=[100,200,300,400,500]
fig = plt.figure(figsize=(20,12))

for n in range(len(Psi)):
    plt.subplot(2,3,n+1)
    plt.xlabel("t,($10^{-18} s$)",fontsize=12)
    plt.ylabel("$\psi(t)$")               
    plt.title('Probability amplitude of $\psi(t)$ at '+str(Psi[n])+'steps',fontsize=15)
    plt.plot(psi_t[Psi[n]-1].real)
```

Extend your program further to make an animation of the wavefunction over time, displaying the real part at each time step

```python
## For animation:
start=time()

psi_animate=np.empty((5000,N),complex) #Calculating psi values for successive timesteps and storing in a new array
psi_animate[0]=psi

for i in range(1,len(psi_animate)):
    v=np.matmul(B,psi_animate[i-1])
    psi_animate[i]=banded(A_band,v,1,1)
    
    psi_animate[:,0]=psi_animate[:,N-1]=0    ### Boundary condition on psi
    
end=time()
print( 'Run time for this cell is', end-start,'seconds')
```

```python

from IPython.display import HTML
from matplotlib.animation import FuncAnimation

# creating a blank window
# for the animation
fig = plt.figure()
axis = plt.axes(xlim =(0, N),ylim =(-1, 1))
 
line, = axis.plot([], [], lw = 2)
 
# what will our line dataset
# contain?
def init():
    line.set_data([], [])
    return line,

# animation function
def animate(i):
    x = np.linspace(0, N, N)
    y = ((psi_animate[i].real))  
    line.set_data(x, y)
    return (line,)
 
# calling the animation function    
anim = FuncAnimation(fig, animate,init_func = init,frames = 5000,interval = 10, blit = True)

# anim.save('Wave_function.mp4', writer = 'ffmpeg', fps = 100)

HTML(anim.to_html5_video())
```
