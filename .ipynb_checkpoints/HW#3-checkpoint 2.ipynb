# Problem#1:Newman 5.13, modified: Kinetic energy of the quantum harmonic oscillator

Defining the function:
$$\psi_n(x)=\frac{1}{\sqrt{2^nn!\sqrt \pi}}e^{-x^2/2}H_n(x)$$

```python
def psi(x,n):
    if n==0:
        return 1/(np.sqrt(2**n*np.sqrt(np.pi)))*np.exp(-x**2/2)
    elif n==1:
        return 1/(np.sqrt(2**n*np.sqrt(np.pi)))*np.exp(-x**2/2)*2*x
    else:
        None
```

```python
fig = plt.figure(figsize=(10,10))
x_axis = []
psi_0x = []
psi_1x = []
for x in np.linspace(-4,4,100):
    x_axis.append(x)
    psi_0x.append(psi(x,0))
    psi_1x.append(psi(x,1))
# print(y_CV_mm)
plt.plot(x_axis,psi_0x,label='$\psi_0(x)$')
plt.plot(x_axis, psi_1x,label='$\psi_1(x)$')
plt.xlabel(" $x$ ",fontsize=15)
plt.ylabel(" $\psi_n(x)$ ",fontsize=15)
plt.legend(loc='best',fontsize=15)
plt.title('Harmonic oscillator wavefunctions  $\psi_n(x)$ ')
plt.show()
```
