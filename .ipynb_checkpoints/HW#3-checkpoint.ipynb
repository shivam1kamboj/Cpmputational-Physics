# Problem#1:Newman 5.13, modified: Kinetic energy of the quantum harmonic oscillator

Defining the function:
$$\psi_n(x)=\frac{1}{\sqrt{2^nn!\sqrt \pi}}e^{-x^2/2}H_n(x)$$

```python
def psi(x,n):
    if n==0:
        return 1/(np.sqrt(2**n*np.sqrt(np.pi)))*np.exp(-x**2/2)
    elif n==1:
        return 1/(np.sqrt(2**n*np.sqrt(np.pi)))*np.exp(-x**2/2)*2*x
    else:
        None
```

```python
x_axis = []
psi_0x = []
psi_1x = []
for x in np.linspace(-4,4,100):
    x_axis.append(x)
    psi_0x.append(psi(x,0))
    psi_1x.append(psi(x,1))
fig = plt.figure(figsize=(10,10))
plt.plot(x_axis,psi_0x,label='$\psi_0(x)$')
plt.plot(x_axis, psi_1x,label='$\psi_1(x)$')
plt.xlabel(" $x$ ",fontsize=15)
plt.ylabel(" $\psi_n(x)$ ",fontsize=15)
plt.legend(loc='best',fontsize=15)
plt.title('Harmonic oscillator wavefunctions:  $\psi_n(x)$ ')
plt.show()
```

$\textbf{Part (b):}$\
Second order derivative is defined as:
$$f''(x)=lim_{h\to 0}\frac{f'(x+h) -f'(x)}{h}$$
which implies: $$f''(x)=lim_{h\to 0}\frac{\frac{f(x+h+h)-f(x+h)}{h} -\frac{f(x+h)-f(x)}{h}}{h}$$
Therefore, $$f''(x)=lim_{h\to 0}\frac{f(x+2h)-2f(x+h)+f(x)}{h^2}$$
$-\frac{1}{2}\frac{d^2}{dx^2}\psi_n(x)$ is defined as:

```python
def Kinetic_psi(x,h,n):
    if n==0:
        return -0.5*(psi((x+2*h),0)-2*psi((x+h),0)+psi(x,0))/h**2
    elif n==1:
        return -0.5*(psi((x+2*h),1)-2*psi((x+h),1)+psi(x,1))/h**2
    else:
        return None
```

```python
x_d=[]
y_0=[]
y_1=[]
h=8/100
for x in np.linspace(-4,4,100):
    x_d.append(x)
    y_0.append(Kinetic_psi(x,h,0))
    y_1.append(Kinetic_psi(x,h,1))
fig = plt.figure(figsize=(10,10))
plt.plot(x_d,y_0,label='$-(1/2)d^2/dx^2\psi_0(x)$')
plt.plot(x_d,y_1,label='$-(1/2)d^2/dx^2\psi_1(x)$')
plt.xlabel(" $x$ ",fontsize=15)
plt.ylabel('$(1/2)d^2/dx^2\psi_n(x)$',fontsize=15)
plt.legend(loc='best',fontsize=15)
plt.title('Kinetic Energy: $-(1/2)d^2/dx^2\psi_n(x)$')
plt.show()
```

$\textbf{Part(c)}$
Energy for Qunatum harmonic ocilator is given by $(n+1/2)$. Which means we have only two cases of energy now:
1. For $n=0$, $E=1/2$
2. For $n=1$, $E=3/2$
 

```python
def delta(n,h,x):
    if n==0:
        return Kinetic_psi(x,h,0)+0.5*(x**2)*(psi(x,0))-1/2*(psi(x,0))
    elif n==1:
        return Kinetic_psi(x,h,1)+0.5*(x**2)*(psi(x,1))-3/2*(psi(x,1))
    else:
        return None
```

Iteration to find out the minimum value of h for which Delta gives the minimum energy difference for $\Delta_0(x)$

```python
###This loop gives me the value of h after which energy difference will begin to increase
###----------------------------###
###----------------------------###
delta_01=delta(0,8/100,np.linspace(-8/1000,16/1000,3)) ##My first Delta value for psi(0,x)
h=8/1000
delta_02=delta(0,h,np.linspace(-h,2*h,3))     ##My second Delta value for psi(0,x)

while (abs(delta_01)>abs(delta_02)).all():     #Whenever Second Delta value become greater than first break the loop
    delta_01=delta(0,h,np.linspace(-h,2*h,3))
    h*=0.25
    delta_02=delta(0,h,np.linspace(-h,2*h,3))
    print('Delta_0 is ',abs(delta_02[1]),'for h =',h)
```

Iteration to find out the minimum value of h for which Delta gives the minimum energy difference for $\Delta_1(x)$

```python
delta_11=delta(1,8/100,np.linspace(-8/100,16/100,3))
h=8/1000
delta_12=delta(1,h,np.linspace(-h,2*h,3))
while (abs(delta_11)>abs(delta_12)).all():
    delta_11=delta_12
    h*=0.25
    delta_12=delta(1,h,np.linspace(-h,2*h,3))
    print('Delta_1 is',abs(delta_12[1]),'for h =',h)   
```

# So after $h=1.25\times10^{-5}$ if I decrease h, the energy difference $\Delta_0(x)$ begin to increase. 

```python
fig = plt.figure(figsize=(25,10))
plt.subplot(1,2,1)
for h in [0.005,0.000125,3.125e-05]:
    x_delta=[]
    delta_0=[]
    for x in np.linspace(-4,4,100):
        x_delta.append(x)
        delta_0.append(delta(0,h,x))
    plt.plot(x_delta,delta_0,label="$\Delta_0(x)$ for h="+str(h))
plt.legend(loc='best',fontsize=15)
plt.xlabel(" $x$ ",fontsize=15)
plt.ylabel('$\Delta_0(x)$',fontsize=15)
plt.title('Energy difference $\Delta_0(x)$')

plt.subplot(1,2,2)
for h in [1.220703125e-07,3.0517578125e-08,7.62939453125e-09]:
    x_delta=[]
    delta_1=[]
    for x in np.linspace(-4,4,100):
        x_delta.append(x)
        delta_1.append(delta(1,h,x))
    plt.plot(x_delta,delta_1,label="$\Delta_1(x)$ for h="+str(h))
plt.legend(loc='best',fontsize=15)
plt.xlabel(" $x$ ",fontsize=15)
plt.ylabel('$\Delta_1(x)$',fontsize=15)
plt.title('Energy difference $\Delta_1(x)$')
plt.show()
```

Energy differnce $\Delta_1(x)$ for smaller values of h is significantly bigger than $\Delta_0(x)$. Now I am plotting both $\Delta_0(x)$ and $\Delta_1(x)$ on the same plot for $h=3.125\times10^{-5}$ and $h=5\times10^{-5}$ because for these values 

```python
fig = plt.figure(figsize=(10,10))
for h in [5e-05,3.125e-05]:
    x_delta=[]
    delta_0=[]
    delta_1=[]
    for x in np.linspace(-4,4,100):
        x_delta.append(x)
        delta_0.append(delta(0,h,x))
        delta_1.append(delta(1,h,x))
    print("Maximum value of Delta_0(x)=",max(delta_0),"for h=",h)
    print("Maximum value of Delta_1(x)=",max(delta_1),"for h=",h)
    plt.plot(x_delta,delta_0,label="$\Delta_0(x)$ for h="+str(h))
    plt.plot(x_delta,delta_1,label="$\Delta_1(x)$ for h="+str(h))
plt.legend(loc='best',fontsize=15)
plt.xlabel(" $x$ ",fontsize=15)
plt.ylabel('$\Delta_n(x)$',fontsize=15)
plt.title('$\Delta_n(x)$ for n=0, and 1')
plt.show()
```

# Conclusion: $h=5\times10^{-5}$ is best stepsize for computing kinetic energy for $\psi_n(x)$ for both $n=0$ and $n=1$

# Problem:#2 Newman 6.4: A circuit of resistors, with numpy

$\textbf{Part a}$\
According to krichoff's law:
1. At junction V_1
$$\frac{V_1-V_2}{R}+\frac{V_1-V_3}{R}+\frac{V_1-V_4}{R}+\frac{V_1-V_+}{R}=0$$
0r $$4V_1-V_2-V_3-V_4=5$$
2. At junction V_2
$$\frac{V_2-V_1}{R}+\frac{V_2-V_4}{R}+\frac{V_2-0}{R}=0$$
Or $$3V_2-V_1-V_4=0$$
3. At junction V_3
$$\frac{V_3-V_1}{R}+\frac{V_3-V_4}{R}+\frac{V_3-V_+}{R}=0$$
Or $$3V_3-V_1-V_4=5$$
4. At junction V_4
$$\frac{V_4-V_1}{R}+\frac{V_4-V_2}{R}+\frac{V_4-V_3}{R}+\frac{V_4-0}{R}=0$$
Or $$4V_4-V_1-V_2-V_3=0$$

```python
Voltage_matrix=np.array([[4,-1,-1,-1],[-1,3,0,-1],[-1,0,3,-1],[-1,-1,-1,4]])
print(Voltage_matrix)
Given_Voltage=np.array([5.0,0,5.0,0])
Given_Voltage
```

$\textbf{Part b}$\
Solving the matrix using np.linalg.solve to find out Voltages $V_1$, $V_2$, $V_3$, and $V_4$


```python
V=np.linalg.solve(Voltage_matrix,Given_Voltage)
print('V_1,V_2,V_3,V_4=',V)
len(V)
```

$\textbf{Part c}$\
Verifying the results obtained if the voltages satisfy the Krichoff's law equations on each juction.

```python
Result=np.matmul(Voltage_matrix,V)
##Let's check:
Result==Given_Voltage
```

To see the reason let's print the Result and see what else can be done to make the result true

```python
print(Result)
```

So it means the error is due to round off problem. So let's round of the numbers using python round function

```python
Round_of_Voltage=[]
for i in range(len(Result)):
    Result[i]=round(Result[i])
    Round_of_Voltage.append(Result[i])
type(Round_of_Voltage)
Rounded_Result=np.array(Round_of_Voltage)
```

```python
##Let's check again:
Given_Voltage==Rounded_Result
```

```python
print(Rounded_Result)
```

The above vector Rounded_Result is the right hand side of Krichoff's equations we wrote for each junction i.e at $V_1$, $V_2$, $V_3$, and $V_4$ were 5,0,5, and 0 respectively

# Now, defining and computing partial derivates $\partial{w}/\partial{x}$, and $\partial{w}/\partial{y}$ of the given data(Image)

```python
def delw_dely(data,dimension,h):
    partial_y=np.zeros_like(data)
    partial_y[0,:]=(data[1,:]-data[0,:])/h
    for i in range(1,dimension):
        partial_y[i,:]=(data[i,:]-data[i-1,:])/h
    partial_y[dimension-1,:]=(data[dimension-1,:]-data[dimension-2,:])/h
    return partial_y
        
def delw_delx(data,dimension,h):
    partial_x=np.zeros_like(data)
    partial_x[:,0]=(data[:,1]-data[:,0])/h
    width=len(data[0,:])   ### because in x-direction, array has 1024 columns which I took as width of Image
    for i in range(1,width):
        partial_x[:,i]=(data[:,i]-data[:,i-1])/h
    partial_x[:,width-1]=(data[:,width-1]-data[:,width-2])/h
    return partial_x
```

```python
partial_w_x=delw_delx(Image,512,30000)
partial_w_y=delw_dely(Image,512,30000)
print('shape of del_w/x=',np.shape(partial_w_x))
print('shape of del_w/y=',np.shape(partial_w_y))
```

$\textbf{Part b}$\
$\textbf{Defining and then calculating the intensity for each grid using}$\
\
$$I = \frac{cos\phi(\partial{w}/\partial{x})+sin\phi(\partial{w}/\partial{y})}{\sqrt{(\partial{w}/\partial{x})^2+(\partial{w}/\partial{y})^2+1}}$$

```python
def Intensity(phi,partial_x,partial_y):
    return (np.cos(phi)*partial_x+np.sin(phi)*partial_y)/(np.sqrt(partial_x**2+partial_y**2+1))
```

```python
I=Intensity(np.pi/4,partial_w_x,partial_w_y)
np.shape(I)
```
