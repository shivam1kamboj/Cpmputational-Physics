## Question 1

### 1a) 

\begin{align}
E_1 = \frac{1}{2}mv_1^2 - G \frac{mM}{\ell_1} \\
E_2 = \frac{1}{2}mv_2^2 - G \frac{mM}{\ell_2} \\
E_1 = E_2 \\
\frac{1}{2}mv_1^2 - G \frac{mM}{\ell_1} = \frac{1}{2}mv_2^2 - G \frac{mM}{\ell_2} \\
v_1^2 - \frac{2 G M}{\ell_1} = v_2^2 - \frac{2GM}{\ell_2}
\end{align}
Then use $v_1 \ell_1 = v_2 \ell_2$:
\begin{align}
v_1^2 - \frac{2 G M}{\ell_1} = v_2^2 - \frac{2GM}{v_1 \ell_1} v_2 \\
v_2^2 - \frac{2GM}{v_1 \ell_1} v_2 - \left[ v_1^2 - \frac{2 G M}{\ell_1} \right] = 0
\end{align}

This is a quadratic equation with two roots. One root is $v_2 = v_1$, clearly, but that is the perihelion. The other root is the aphelion, where the distance is larger and the velocity is smaller.

### 1b)

```python
from numpy import sqrt, pi

MM = 1.9891e30
# kg, mass of Sun
GG = 6.6738e-11
# m^3/kg-s^2, Newton's gravitational constant

def semi_major(l1, l2):
    return (l1+l2)/2

def semi_minor(l1, l2):
    return sqrt(l1*l2)

def orbital_period(l1, l2, v1):
    return 2*pi*semi_major(l1,l2)*semi_minor(l1,l2)/(l1*v1)

def eccentricity(l1, l2):
    return (l2-l1)/(l2+l1)

def velocity_aphelion(l1, v1):
# (-b - sqrt(b^2-4ac))/2a
    bb = 2*GG*MM/(v1*l1)
    return (bb-sqrt(bb**2 + 4*(v1**2-2*GG*MM/l1)))/2

def total_energy(ll, vv):
    return vv**2/2 - GG*MM/ll

def calc_parameters(l1, v1):
#    l1 = float(input("Enter distance to the Sun at perihelion:"))
#    v1 = float(input("Enter velocity at perihelion:"))
    v2 = velocity_aphelion(l1, v1)
    l2 = l1*v1/v2
    
    # these lines were not requested but can be a useful check for proper quadratic solution
    print("Total energy at perihelion: {:e}".format(total_energy(l1, v1)), "J / kg")
    print("Total energy at aphelion:   {:e}".format(total_energy(l2, v2)), "J / kg")
    
    print("Distance to the Sun at aphelion (l2): {:e}".format(l2), "m")
    print("Velocity at aphelion (v2): {:e}".format(v2), "m/s")
    print("Semi-major axis: {:e}".format(semi_major(l1, l2)), "m")
    print("Semi-minor axis: {:e}".format(semi_minor(l1, l2)), "m")
    print("Orbital period: {:e}".format(orbital_period(l1, l2, v1)), "s =",
          orbital_period(l1, l2, v1)/(365.25*24*60*60), "y")
    print("Orbital eccentricity: ", eccentricity(l1, l2))
```

### 3a)

```python
from pylab import imshow, figure, gray
from numpy import zeros

maxiter = 100

def iterate(xx,yy):
    cc = complex(xx, yy)
    zz = 0 + 0j
    for ii in range(1,maxiter):
        zz = zz**2 + cc
        if(abs(zz) > 2.0):
            return 1
    return 0

NN = 1000

# map integers 1 to NN into -2 to 2
def coord(xx):
    return 4*(xx/NN-0.5)

mandelbrot = zeros([NN,NN],int)
for xx in range(1,NN):
    for yy in range(1,NN):
        mandelbrot[xx,yy] = iterate(coord(xx),coord(yy))
               
figure(num=1, figsize=(10, 10))
gray()
imshow(mandelbrot, extent=[-2,2,-2,2])
# the default coloration ends up giving exactly as specified in the question
```

### 3b)

```python
from pylab import imshow, figure, hot
from numpy import zeros

maxiter = 100

def iterate(xx,yy):
    cc = complex(xx, yy)
    zz = 0 + 0j
    for ii in range(1,maxiter):
        zz = zz**2 + cc
        if(abs(zz) > 2.0):
            return ii
    return maxiter

NN = 1000

# map integers 1 to NN into -2 to 2
def coord(xx):
    return 4*(xx/NN-0.5)

mandelbrot = zeros([NN,NN],int)
for xx in range(1,NN):
    for yy in range(1,NN):
        mandelbrot[xx,yy] = iterate(coord(xx),coord(yy))
               
figure(num=1, figsize=(10, 10))
hot()
imshow(mandelbrot, extent=[-2,2,-2,2])
```

### 3c)

```python
from pylab import imshow, figure, jet
from numpy import zeros, log

# turns out at this size, you don't get much more detail for maxiter > 100 or NN > 1000

maxiter = 100

def iterate(xx,yy):
    cc = complex(xx, yy)
    zz = 0 + 0j
    for ii in range(1,maxiter):
        zz = zz**2 + cc
        if(abs(zz) > 2.0):
            return log(ii)
    return log(maxiter)

NN = 1000

# map integers 1 to NN into -2 to 2
def coord(xx):
    return 4*(xx/NN-0.5)

mandelbrot = zeros([NN,NN],int)
for xx in range(1,NN):
    for yy in range(1,NN):
        mandelbrot[xx,yy] = iterate(coord(xx),coord(yy))
               
figure(num=1, figsize=(10, 10))
jet()
imshow(mandelbrot, extent=[-2,2,-2,2])
```
