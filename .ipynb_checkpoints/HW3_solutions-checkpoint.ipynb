# Question 1

# 1a)

$$\psi_0 (x) = \frac{1}{\pi^{1/4}}e^{-x^2/2}, \psi_1 (x) = \frac{\sqrt{2}}{\pi^{1/4}}e^{-x^2/2} x$$
Compare: https://en.wikipedia.org/wiki/Hermite_polynomials#Hermite_functions

```python
def psi0(x):
    return exp(-x**2/2)/(pi**4)

def psi1(x):
    return sqrt(2)*x*exp(-x**2/2)/(pi**4)

minx = -4
maxx = 4

def calc_psis():
    xpts = linspace(minx, maxx, npts)
    psi0pts = empty([npts], float)
    psi1pts = empty([npts], float)
    for ii in range(0,npts):
        psi0pts[ii] = psi0(xpts[ii])
        psi1pts[ii] = psi1(xpts[ii])
    return xpts, psi0pts, psi1pts
        
def plot_psis():
    plot(xpts, psi0pts, '-', label="$\Psi_0$")
    plot(xpts, psi1pts, '-', label="$\Psi_1$")
    xlabel("x")
    ylabel("$\Psi(x)$")
    legend()
    show()
```

```python
npts = 100
xpts, psi0pts, psi1pts = calc_psis()
plot_psis()
```

# 1b)

```python
def d2dx2(psi):
    hh = (maxx-minx)/npts
    d2dx2pts = empty([npts], float)
    for ii in range(1,npts-1):
        d2dx2pts[ii] = (psi(xpts[ii-1]) - 2*psi(xpts[ii]) + psi(xpts[ii+1]))/(hh**2)   
    
    # or do it with arrays as below, for greater speed
    #d2dx2pts[1:npts-1] = psi(xpts[0:npts-2]) - 2*psi(xpts[1:npts-1]) + psi(xpts[2:npts])/(hh**2)
    
    # set edges to zero as they cannot be calculated by this formula
    d2dx2pts[0] = 0
    d2dx2pts[npts-1] = 0
    return d2dx2pts

def calc_d2dx2():
    dpsi0dx2pts = d2dx2(psi0)
    dpsi1dx2pts = d2dx2(psi1)
    return dpsi0dx2pts, dpsi1dx2pts

def plot_d2dx2():
    plot(xpts, dpsi0dx2pts, '-', label="$\Psi''_0$")
    plot(xpts, dpsi1dx2pts, '-', label="$\Psi''_1$")
    xlabel("x")
    ylabel("$\Psi''(x)$")
    legend()
    show()
```

```python
dpsi0dx2pts, dpsi1dx2pts = calc_d2dx2()
plot_d2dx2()
```

# 1c)

```python
def calc_delta(energy, psipts, dpsidx2pts):
    delta = empty([npts], float)
    delta[:] = -0.5*dpsidx2pts[:] + (0.5*(xpts[:]**2) - energy)*psipts[:]

    # just set to zero on the edges as we were not able to calculate the kinetic energy
    delta[0] = 0
    delta[npts-1] = 0
    return delta

def plot_deltas():
    delta0 = calc_delta(0.5, psi0pts, dpsi0dx2pts)
    delta1 = calc_delta(1.5, psi1pts, dpsi1dx2pts)
    plot(xpts, delta0, '-', label="$\Delta_0$")
    plot(xpts, delta1, '-', label="$\Delta_1$")
    xlabel("x")
    ylabel("$\Delta$")
    legend()
    show()
```

```python
plot_deltas()
```

```python
npts = 1000
xpts, psi0pts, psi1pts = calc_psis()
dpsi0dx2pts, dpsi1dx2pts = calc_d2dx2()
plot_deltas()
```

We varied $N$, with $h = 8/N$. $\Delta$ goes from $\sim 10^{-3}$ at $h=8/20$, to $\sim 10^{-5}$ at $h=8/100$ or $8/1000$, to $\sim 10^{-7}$ at $h = 8/10^5$ or $8/10^6$. The last one is extremely noisy. We aren't able to do better than this because of loss of precision from round-off error becoming dominant, even though the approximation error would still be going down. $h=8/10^7$ becomes worse again. Note that we could make the calculation faster, if desired, by just decreasing $h$ and keeping $N$ fixed.

# 2a)

Since all resistors have the same resistance $R$, we can just multiply through by $R$ in all equations. Then we have:
\begin{align}
\left( V_1 - V_2 \right) + \left( V_1 - V_3 \right) + \left( V_1 - V_4 \right) + \left( V_1 - V_+ \right) = 0 \\
\left( V_2 - V_1 \right) + \left( V_2 - V_4 \right) + \left( V_2 - 0 \right) = 0 \\
\left( V_3 - V_1 \right) + \left( V_3 - V_4 \right) + \left( V_3 - V_+ \right) = 0 \\
\left( V_4 - V_1 \right) + \left( V_4 - V_2 \right) + \left( V_4 - V_3 \right) + \left( V_4 - 0 \right) = 0
\end{align}

Simplifying:
\begin{align}
4 V_1 - V_2 - V_3 - V_4 - V_+ = 0 \\
3 V_2 - V_1 - V_4 = 0 \\
3 V_3 - V_1 - V_4 - V_+ = 0 \\
4 V_4 - V_1 - V_2 - V_3 = 0
\end{align}
Note that $V_+$ = 5 V.

or in matrix form:
\begin{align}
\begin{pmatrix}
 4 & -1 & -1 & -1 \\
-1 &  3 &  0 & -1 \\
-1 &  0 &  3 & -1 \\
-1 & -1 & -1 &  4
\end{pmatrix}
\begin{pmatrix}
V_1 \\
V_2 \\
V_3 \\
V_4
\end{pmatrix}
=
\begin{pmatrix}
5 \\
0 \\
5 \\
0
\end{pmatrix}
\end{align}

```python
from numpy.linalg import solve
from numpy import array

A = array([[ 4, -1, -1, -1 ],
           [-1,  3,  0, -1 ],
           [-1,  0,  3, -1 ],
           [-1, -1, -1,  4 ]], float)
v = array([ 5, 0, 5, 0 ],float)

x = solve(A, v)
print(x)
```

# 2c)

```python
from numpy import matmul
print(matmul(A,x))
```

Agrees with the desired vector (5, 0, 5, 0).

```python
from numpy import copy, pi, cos, sin, sqrt

def calc_intensity(data, phi, h):
    
    N = len(data[0,:])
    M = len(data[:,0])
    print("Data is ", N, " by ", M)
    
    intens = empty([M,N], float)
    dwdx = empty([M,N], float)
    dwdy = empty([M,N], float)
    
    # This is a way we could do central differences in the center with a loop instead
#    for y in range(1,N-1):
#        for x in range(1,M-1):
#            dwdx = (data[x+1,y] - data[x-1,y])/(2*h)
#            dwdy = (data[x,y+1] - data[x,y-1])/(2*h)
#            intens[x,y] = (cos(phi)*dwdx + sin(phi)*dwdy)/sqrt(dwdx**2 + dwdy**2 + 1)

    # Use of array operations as below is much faster

    # central differences for x derivative, except at x edges, where one of the two points doesn't exist
    dwdx[1:M-1,:] = (data[2:M,:] - data[0:M-2,:])/(2*h)
    # central differences for y derivative, except at y edges, where one of the two points doesn't exist
    dwdy[:,1:N-1] = (data[:,2:N] - data[:,0:N-2])/(2*h)    

    # left edge, forward difference
    dwdx[0,:] = (data[1,:] - data[0,:])/h
    # right edge, backward difference
    dwdx[M-1,:] = (data[M-1,:] - data[M-2,:])/h

    # bottom edge, forward difference
    dwdx[:,0] = (data[:,1] - data[:,1])/h
    # right edge, backward difference
    dwdx[:,N-1] = (data[:,N-1] - data[:,N-2])/h

    # for the map, another option would be to use periodic boundary conditions in the x-direction,
    # since the Earth is round
    
    intens = (cos(phi)*dwdx + sin(phi)*dwdy)/sqrt(dwdx**2 + dwdy**2 + 1)
    
    return intens
```

# 3b)

```python
intens = calc_intensity(data, pi/4, 30000)
figure(num=1, figsize=(15, 15))
gray()
imshow(intens, vmin=-0.02, vmax=0.02)
colorbar()
```
